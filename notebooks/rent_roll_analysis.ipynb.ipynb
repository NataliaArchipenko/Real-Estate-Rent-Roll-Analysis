# Real Estate Rent Roll Analysis

## Project Overview

This project analyzes a **commercial real estate rent roll dataset** to gain insights into property occupancy, lease expirations, tenant distribution, and potential risks.

A **rent roll dataset** typically contains detailed information about lease agreements within real estate properties, including tenant data, leased area, lease terms, and property information.

The goal of this analysis is to answer the following business questions:

- What is the occupancy rate of each property?
- Which leases are expiring soon?
- How is the rentable area distributed by usage type (e.g., office, warehouse, production)?
- Are there risks due to tenant concentration?
- Which properties show vacancy or underutilization?

## Data Cleaning

Before performing the analysis, we need to clean the dataset.

The following steps will be performed:

1. Remove unnecessary columns
2. Convert date columns
3. Handle missing values
4. Check duplicates
5. Prepare variables for analysis

```python
# Drop unnecessary columns

columns_to_drop = [
    "Spalte1",
    "Check_Area",
    "Check_Rent",
    "Check"
]

df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

df.head()
```

## Convert Date Columns

Lease start and end dates are important for analyzing lease durations and expirations.

We convert the date columns into proper datetime format.

```python
# Convert date columns (adjust column names if necessary)

df["Lease_Start"] = pd.to_datetime(df["Lease_Start"], errors="coerce")
df["Lease_End"] = pd.to_datetime(df["Lease_End"], errors="coerce")

df[["Lease_Start", "Lease_End"]].head()
```

```python
# Check missing values

missing_values = df.isnull().sum()

missing_values.sort_values(ascending=False)
```

Some columns contain missing values, particularly those related to lease information and calculated metrics.

These columns will be reviewed carefully to determine whether the missing values represent vacant units, incomplete records, or data entry issues.

## Calculate Lease Duration

We calculate the lease duration in years to better understand the contract structure.

```python
# Calculate lease duration in years

df["Lease Duration Years"] = (df["Lease_End"] - df["Lease_Start"]).dt.days / 365

df["Lease Duration Years"].head()
```
